```python
import pandas as pd

# Dados da tabela
dados = {
    "Meses": ["Abril", "Maio", "Junho", "Julho"],
    "Perdas": [176543, 189021, 198765, 182123]
}

df = pd.DataFrame(dados)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df["Meses"], df["Perdas"])
plt.title("Perdas por Mês (2020)")
plt.xlabel("Meses")
plt.ylabel("Total de Perdas")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8,5))

sns.lineplot(data=df, x="Meses", y="Perdas", marker="o")
plt.title("Evolução das Perdas entre Abril e Julho (2020)")
plt.xlabel("Meses")
plt.ylabel("Total de Perdas")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(df["Perdas"], labels=df["Meses"], autopct="%1.1f%%")
plt.title("Proporção das Perdas por Mês")
plt.show()
```
