Carlos é dono de um restaurante que possui filiais em algumas cidades do Brasil e deseja verificar a quantidade de clientes durante o mês de dezembro por cada filial. Como Carlos pode analisar essa informação da forma mais simples?



Para isso, use comandos Python.



Tabela 1 – Quantidade de clientes por filial

São Paulo 1.034

Rio de Janeiro 2.345

Fortaleza 678

Natal 456

```python
#montar o dataframe

import pandas as pd

#texto copiado da página do exercício

texto = """São Paulo

1.034

Rio de Janeiro

2.345

Fortaleza

678

Natal

456"""

# separa linhas e remove vazias
linhas = [l.strip() for l in texto.split("\n") if l.strip()]

# cidades estão nas posições pares, valores nas ímpares
cidades = linhas[0::2]
valores = linhas[1::2]

df = pd.DataFrame({
    "Cidade": cidades,
    "Valor": valores
})

#converte valores para números
df["Valor"] = df["Valor"].str.replace(".", "", regex=False).astype(int)
print(df)

```

```python
import matplotlib.pyplot as plt

plt.barh(df["Cidade"], df["Valor"])
plt.show()
```
