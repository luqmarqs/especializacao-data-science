```python
#Inicialmente extrairemos a média apenas de uma coluna: sepal_width.
media = np.mean(data["sepal_width"])
print(media)
```

```python
# media de todas as colunas:
media_geral = np.mean(data)
print(media_geral)
```

```python
#Pode-se extrair a mediana dos dados através do comando np.median()
mediana = np.median(data["sepal_length"])
print(mediana)
```

```python
#Para inferir sobre a moda no conjunto de dados, podemos usar o comando .mode()
moda = data["sepal_length"].mode()
print(moda)
```

```python
# PERCENTIS E QUARTIS
# Para a análise de 10% dos dados:
P10 = np.quantile(data["sepal_width"], 0.10)
print(P10)
```

```python
#Para analisar o primeiro quartil temos:
Q1 = np.quantile(data["sepal_width"], 0.25)
print(Q1)
```

```python
#Para fazer um resumo da estatística descritiva dos dados, usamos o comando .describe().
data.describe()
```

```python
#Antes de calcularmos a amplitude dos dados, precisamos avaliar os valores máximo e mínimo do conjunto
data.max()
```

```python
data.min()
```

```python
#Como temos valores não numéricos no conjunto de dados “virginica” e “setosa”, por enquanto, não podemos fazer o cálculo da amplitude. No caso, precisamos remover as colunas com valores não numéricos e, para isso, podemos usar o comando .iloc[].
Amplitude = data.iloc[:, 0:4].max() - data.iloc[:, 0:4].min()
print(Amplitude)
```

```python
#Para calcular as medidas de dispersão variância e desvio padrão no Python, seguimos com os seguintes comandos:

#Variância
np.var(data)
```

```python
#Desvio padrão
np.std(data)
```

```python
import seaborn as sns
%matplotlib inline

#O Seaborn contém alguns conjuntos de dados pré-carregados bastante utilizados para estudo e pesquisa. 
#Podemos ter acesso através do comando:
data = sns.load_dataset("tips")
```

```python
#Esse é um dataset sobre a gorjeta.
#Inicialmente, vamos analisar o total de contas “total_bill”.
sns.distplot(data["total_bill"])
```

```python
#Agora verificaremos a quantidade de gorjeta por sexo:
sns.barplot(x='sex', y='tip', data=data)
```

```python
#Bom, analisaremos as gorjetas por dia através de um boxplot:
sns.boxplot(x='day', y='tip', data=data)
```

```python
#Por fim, avaliaremos a relação entre as variáveis do banco:
sns.pairplot(data)
```
